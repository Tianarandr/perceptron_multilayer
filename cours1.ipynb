```python
import keras
import numpy as np
from keras.datasets import mnist
from keras.layers import Dense, Dropout, Flatten
from keras.models import Sequential
from keras.models import Sequential
from keras.utils import np_utils

```

```python
# 1 LOAD DATA FROM MNIST ATASET
(x_train, y_train), (x_test, y_test) = mnist.load_data()
```

```python
np.shape(x_train)
np.shape(x_test)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(7.195, 3.841), dpi=100)
for i in range(100):
  plt.subplot(10,10,i+1)
  plt.imshow(x_train[i].reshape([28, 28]), cmap='gray')
  plt.axis('off')
```

```python
x_train = x_train.reshape(60000, 28*28)
x_test = x_test.reshape(10000, 28*28)
```

```python
x_train = x_train.astype('float32')
x_test = x_test.astype('float32')

#normaliser les donnes
x_train /= 255
x_test /= 255

print(x_train.shape[0], 'train samples')
print(x_test.shape[0], 'train samples')

y_train


```

One hot encoding

```python
#convert class vectors to binary class images
Y_train = np_utils.to_categorical(y_train, 10)
Y_test = np_utils.to_categorical(y_test, 10)

```

```python
np.shape(Y_train)
```

```python
from keras.layers import Dense, Activation
from tensorflow.keras.optimizers import SGD

learning_rate = 0.7
batch_size = 300
nb_epoch = 30

model = Sequential()
model.add(Dense(300, input_dim=784))
model.add(Activation('sigmoid'))

model.add(Dropout(0.3))

model.add(Dense(128))
model.add(Activation('sigmoid'))

model.add(Dropout(0.3))

model.add(Dense(10))
model.add(Activation('softmax'))


sgd = SGD(learning_rate)


model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
#model.compile(loss='categorical_crossentropy', optimizer='adm', metrics=['accuracy'])

model.fit(x_train, Y_train, batch_size=batch_size, epochs=nb_epoch, verbose=1)
```

```python
score = model.evaluate(x_test, Y_test)
```

```python
from sklearn.metrics import confusion_matrix

y_predict = np.argmax(model.predict(x_test), axis=-1)

y_test_labels = Y_test.argmax(1)
print("Confusion Matrix", confusion_matrix(y_test_labels, y_predict))
```
